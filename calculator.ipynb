```python
import math

while(True):
    class Calculator:
        def __init__(self, first, second):
            self.first = first
            self.second = second

        def add(self):
            result = self.first + self.second
            return result
        
        def subtract(self):
            result = self.first - self.second
            return result
        
        def multiply(self):
            result = self.first * self.second
            return result
        
    class UpgradeCal(Calculator):
        def sqrt_root(self):
            result = self.first ** self.second
            return result
        
        def sqrt(self):
            return math.sqrt(self.first), math.sqrt(self.second)
        
        def divide(self):
            if self.second == 0:
                print('zero division error: cannot divide by 0; try again')
                return 0
            else:
                return self.first / self.second
        
        # def modules(self):
        #     if x < 0:
        #         x += -(x * 2)
        #     if y < 0:
        #         y += -(y * 2)
        #     result = self.first + self.second
        #     return result
        
    x = int(input('Enter first number: '))
        
    # if isinstance(x, int) or isinstance(x, float) == True:
    #     print('not number, please try again')
    #     while isinstance(x, int) or isinstance(x, float) == False:
    #         x = input('Enter first number: ')
    # x = int(x)
    
    y = int(input('Enter second number: '))
    
    # if isinstance(y, int) or isinstance(y, float) == True:
    #     print('not number, please try again')
    #     while isinstance(y, int) or isinstance(y, float) == False:
    #         y = input('Enter second number: ')    
    # y = int(y)


    test1 = UpgradeCal(x, y) # test1 = self / x: first / y: second
    
    str = input('Enter operation: ')
    if str == 'add':
        print(test1.add())
    elif str == 'subtract':
        print(test1.subtract())
    elif str == 'multiply':
        print(test1.multiply())
    elif str == 'sqrt_root':
        print(test1.sqrt_root())
    elif str == 'sqrt':
        print(test1.sqrt())
    elif str == 'divide':
        print(test1.divide())
    else:
        print('operator not found')
    reset = input('enter AC to continue and reset inputs')
    
    if reset == 'AC' or reset == 'ac':
        x = 0
        y = 0
    else:
        break
```
